## 課題1

各問で示されている図を出力してください。

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python

def draw_triangle(size):
    # 上から 1 つずつ * を増やしていく
    for i in range(1, size + 1):
        # 1 行分の '*' を組み立て
        line = ""
        for j in range(i):
            line += "*"
        print(line)

if __name__ == "__main__":
    size = 5  
    draw_triangle(size)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
def draw_diagonal(size):
    # size 行の斜めの '*' を描画
    for i in range(size):
        # i 個の空白を積み上げ
        line = ""
        for _ in range(i):
            line += " "
        # その先に '*' を置く
        line += "*"
        print(line)

if __name__ == "__main__":
    size = 5   # ← ここを変えれば行数（斜め線の長さ）を調整できます
    draw_diagonal(size)
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
def draw_rev_diagonal(size):
    # size 行の、右上から左下への斜め線を描画
    for i in range(size):
        line = ""
        # 先頭に (size−1−i) 個の空白
        for _ in range(size - 1 - i):
            line += " "
        # その先に '*'
        line += "*"
        print(line)

if __name__ == "__main__":
    size = 5   # ← ここを変えれば斜め線の長さを調整できます
    draw_rev_diagonal(size)
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
def draw_x(size):
    # size 行・size 列の中で、2 本の対角線を '*' で描画
    for i in range(size):
        line = ""
        for j in range(size):
            # 左上→右下 または 右上→左下 の位置に '*' を置く
            if j == i or j == size - 1 - i:
                line += "*"
            else:
                line += " "
        print(line)

if __name__ == "__main__":
    size = 5   # ← ここを変えれば図形の大きさを変えられます
    draw_x(size)
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
def draw_pyramid(size):
    # 中心揃えの二等辺三角形を描画
    # 行数＝size、最大幅＝2*size-1
    for i in range(size):
        line = ""
        # 左側の空白を (size−1−i) 個積む
        for _ in range(size - 1 - i):
            line += " "
        # '*' を (2*i+1) 個積む
        for _ in range(2 * i + 1):
            line += "*"
        print(line)

if __name__ == "__main__":
    size = 5   # ← ここを変えれば三角形の高さを変更できます
    draw_pyramid(size)
````
